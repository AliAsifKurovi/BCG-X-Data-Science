## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=12)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

The first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]                                  # dataframe named churn containing id and churn status
churn.columns = ['Companies', 'churn']                          
churn_total = churn.groupby(churn['churn']).count()             
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

### Churn Analysis

```python
# Calculate the overall churn rate
overall_churn_rate = client_df['churn'].mean()*100
print(f"Overall Churn Rate: {overall_churn_rate:.2f}% ")
```

```python
# Calculate the overall Retention rate
overall_retention_rate = 100 - client_df['churn'].mean() * 100
print(f"Overall Retention Rate: {overall_retention_rate:.2f}% ")
```

### Churn by Channel Sales

```python
# Calculate churn rate by channel_sales
churn_by_channel = client_df.groupby('channel_sales')['churn'].mean() * 100

# Display the churn rates by channel_sales
print(churn_by_channel)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=churn_by_channel.index, y=churn_by_channel.values)
plt.title('Churn Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

### Churn by has_gas

```python
# Calculate churn rate by has_gas
churn_by_gas = client_df.groupby('has_gas')['churn'].mean() * 100

# Display the churn rates by has_gas
print(churn_by_gas)
```

```python
plt.figure(figsize=(4, 4))
plt.pie(churn_by_gas, labels=churn_by_gas.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title('Churn Rate by Gas Subscription')
plt.show()
```

### Churn By Origin UP

```python
# Calculate churn rate by num_years_antig
churn_by_num_years = client_df.groupby('num_years_antig')['churn'].mean() * 100

# Display the churn rates by num_years_antig
print(churn_by_num_years)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=churn_by_num_years.index, y=churn_by_num_years.values)
plt.title('Churn Rate by Origin of Subscription')
plt.xlabel('Origin of Subscription')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
